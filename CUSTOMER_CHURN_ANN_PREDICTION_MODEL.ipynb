```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv(r"/content/Churn_Modelling.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Geography"]=df["Geography"].bfill()
df["Age"]=df["Age"].bfill()
df["HasCrCard"]=df["HasCrCard"].bfill()
df["IsActiveMember"]=df["IsActiveMember"].bfill()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop("RowNumber",inplace=True,axis=1)
```

```python
df.head()
```

```python
df.drop(["CustomerId","Surname"],inplace=True,axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
df["Geography"]=df["Geography"].astype("category")
df["Geography"]=df["Geography"].cat.codes

df["Gender"]=df["Gender"].astype("category")
df["Gender"]=df["Gender"].cat.codes
```

```python
x=df.iloc[:,:10]
x
```

```python
y=df.iloc[:,10]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
```

```python
model=Sequential()
model.add(Dense(units=40,activation="relu", input_shape=(x_train.shape[1] ,)))
model.add(Dropout(0.3))

model.add(Dense(units=80,activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=1,activation="sigmoid"))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,batch_size=1000,epochs=60)
```

```python
loss,accuracy=model.evaluate(x_test,y_test.values)
```

```python
pred=model.predict(x_test)
x=[]
for i in np.squeeze(pred):
    x.append(np.rint(i))
predicted=pd.DataFrame(x)
print(pred>0.5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=np.array(x)
x=scaler.fit_transform(x_train)
input_values=(699,0,0,39.0,1,0.00,2,0.0,0.0,93826.63)
input_values=np.array(input_values)
input_values=input_values.reshape(1,-1)
input_values=scaler.transform(input_values)
prediction=model.predict(input_values)
print(prediction)

if prediction>0.5:
    print("Customer will leave the bank")
else:
    print("Customer will not leave the bank")
```
